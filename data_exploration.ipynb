```python
import requests
import time
from multiprocessing import Pool
import json

def pcode_to_data(pcode):
    if int(pcode) % 1000 == 0:
        print(pcode)
    
    page = 1
    results = []
    
    while True:
        try:
            response = requests.get('https://www.onemap.gov.sg/api/common/elastic/search?searchVal={0}&returnGeom=Y&getAddrDetails=Y&pageNum={1}'
                                    .format(pcode, page)).json()
        except requests.exceptions.ConnectionError as e:
            print('Fetching {} failed. Retrying in 2 sec'.format(pcode))
            
            time.sleep(2)
            continue
            
        results = results + response['results']
    
        if response['totalNumPages'] > page:
            page = page + 1
        else:
            break
            
    return results


pool = Pool(processes=5)

postal_codes = range(0, 1000000)
postal_codes = ['{0:06d}'.format(p) for p in postal_codes]

all_buildings = pool.map(pcode_to_data, postal_codes)
all_buildings.sort(key=lambda b: (b['POSTAL'], b['SEARCHVAL']))

jstr = json.dumps([y for x in all_buildings for y in x], indent=2, sort_keys=True)

with open('buildings.json', 'w') as f:
    f.write(jstr.encode('utf-8'))
```

```python
import pandas as pd
import extractPostal
import numpy as np
from multiprocessing import Pool, cpu_count

# Load CSV
df = pd.read_csv('ResaleFlatPrices/Resale-Flat-Prices-1990_1999.csv')

# Define the function to get postal code
def get_postal_code(row):
    street_name, block = row["street_name"], row["block"]
    res = extractPostal.pcode_to_data(street_name)
    
    postal_head = None
    for x in res:
        if x['POSTAL'] != 'NIL':
            postal_head = x['POSTAL'][:3]
            break
    if not postal_head:
        return pd.NA
    
    return str(postal_head) + str(block)

# **Process a chunk of the DataFrame**
def process_chunk(df_chunk):
    return df_chunk.apply(get_postal_code, axis=1)

# **Multiprocessing function**
def parallel_apply(df, num_cores=None):
    if num_cores is None:
        num_cores = cpu_count()  # Use all available CPU cores

    # Split DataFrame into chunks
    df_split = np.array_split(df, num_cores)

    # Use multiprocessing Pool
    with Pool(num_cores) as pool:
        results = pool.map(process_chunk, df_split)  # Calls the global function

    return pd.concat(results)
```

```python
# Apply function in parallel
df["postal_code"] = parallel_apply(df)
```
